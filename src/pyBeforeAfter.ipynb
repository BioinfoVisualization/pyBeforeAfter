```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

## Import dataset

```python
in_file = '../data/treatment_dataset.csv'
```

```python
file_type = in_file.split('.')[-1]
if file_type == 'tsv':
    DF = pd.read_csv(in_file,sep='\t',index_col=0).reset_index(drop=True)
elif file_type == 'xlsx':
    DF = pd.read_excel(in_file,index_col=0)
elif file_type == 'csv':
    DF = pd.read_csv(in_file,sep=',',index_col=0).reset_index(drop=True)
else:
    raise NameError("Invalid input format. Has to be either .tsv, .csv or .xlsx.")
```

## Verification of dataset

```python
timelines = list(DF[time_col].unique())
```

```python
if len(timelines)!=2:
    raise ValueError("Only 2 timelines are allowed, found {}.".format(len(timelines)))
n_b = len(DF.loc[DF[time_col]==timelines[0],time_col])
n_a = len(DF.loc[DF[time_col]==timelines[1],time_col])
if n_b != n_a:
    raise ValueError("The two timepoints must have the same number of samples.")
if t_0 not in timelines:
    raise ValueError('The value specified for first timeline is not present in the dataset.')
timelines.remove(t_0)
t_1 = timelines[0]
```
