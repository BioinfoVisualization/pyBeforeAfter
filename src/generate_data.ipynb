```python
import numpy as np
import pandas as pd
```

```python
N = 36
```

```python
counts_nr = np.random.randint(0,1000,int(N/2))
counts_r0 = np.random.randint(0,1000,int(N/4))
counts_r1 = np.random.randint(0,50,int(N/4))
counts_r = np.array([i for pair in zip(counts_r0,counts_r1) for i in pair])
counts = np.concatenate([counts_r, counts_nr])
```

```python
pat = [x//2 for x in range(N)]
```

```python
response = int(0.5*N)*['Responder']+int(0.5*N)*['Non-responder']
```

```python
timeline = int(0.5*N)*['before','after']
```

```python
DF = pd.DataFrame({'patient':pat,'counts':counts,'response':response,'timeline':timeline})
DF
```

```python
DF.to_csv('../data/treatment_dataset.csv',sep=',')
DF.to_csv('../data/treatment_dataset.tsv',sep='\t')
DF.to_excel('../data/treatment_dataset.xlsx')
```
